```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False 
    return True 

for n in range (1, 201):
    if prime(n):
        print("It is prime!")
    else:
        print("Not a prime!")
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number. of terms.
class Fibonacci:
    def __init__(self, n):
        self.n = n
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            next_value = self.a + self.b
            self.a, self.b = self.b, next_value
            self.count += 1
            return next_value

fib = Fibonacci(10)

for num in fib:
    print(num)
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power(n):
    for i in range(n + 1):
        yield 2 ** i

for value in power(5):
    print(value)
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
#1st we have to create a file in the notebook
with open("sample.txt", "w") as f:
    f.write("Hi\nNext\nBye")

def reads(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line

for line in reads("sample.txt"):
    print(line)
```

```python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
celsius = [0, 50, 15, 95, 4]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print("Temperatures in Fahrenheit:", fahrenheit)
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
def vowel(text):
    vowels = "aeiouAEIOU"
    return "".join(filter(lambda ch: ch not in vowels, text))

s = "Hello World"
result = vowel(s)
print("String without vowels:", result)
```
